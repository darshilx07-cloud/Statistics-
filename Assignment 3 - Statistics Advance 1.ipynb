### Q1. What is a random variable in probability theory?
ANS:- A random variable is a variable whose value is subject to variations due to chance. It represents the
outcomes of a random phenomenon. For example, if you roll a die, the number that comes up is a random
variable, because it’s uncertain until the die is rolled.

### Q2. What are the types of random variables?
ANS:- There are two main types of random variablesT
> Discrete random variables – These take on a finite or countable number of values. An example is the number
of heads when flipping a coin multiple times+
> 
> Continuous random variables – These take on an infinite number of values within a given range. An example
is the height of individuals in a population, which can vary continuously.

### Q3. What is the difference between discrete and continuous distributions?
ANS:- Discrete distributions apply to discrete random variables, where the data points are countable. Examples
include the binomial distribution and Poisson distribution. Continuous distributions, on the other hand, apply
to continuous random variables, where data points can take on any value within a range. Examples include
the normal distribution and the uniform distribution.

### Q4. What are probability distribution functions (PDF)?
ANS:- A probability distribution function (PDF) is a function that describes the likelihood of a random variable
taking on a particular value. For discrete random variables, this function gives the probability of each
possible outcome. For continuous random variables, the PDF gives the likelihood of the random variable
falling within a specific range of values, but it cannot give the probability of a single exact value (as it’s 0 for
any specific point).

### Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
ANS:- The CDF is related to the PDF but provides cumulative probability. While the PDF gives the probability of a
specific outcome, the CDF gives the probability that a random variable is less than or equal to a particular
value. In other words, the CDF sums the probabilities up to that point.

### Q6. What is a discrete uniform distribution?
ANS:- A discrete uniform distribution is a type of distribution where all outcomes are equally likely. For example,
when rolling a fair six-sided die, each face (1 through 6) has an equal probability of appearing, so it follows a
discrete uniform distribution.

### Q7. What are the key properties of a Bernoulli distribution?
ANS:- A Bernoulli distribution models a random experiment with exactly two possible outcomes: success (usually
coded as 1) and failure (coded as 0). The key properties includeT
>It’s a discrete distribution+
>
> The probability of success is denoted by p, and the probability of failure is 1-p.
>
> It has a mean of p and a variance of p(1-p).

### Q8. What is the binomial distribution, and how is it used in probability?
ANS :- The binomial distribution describes the number of successes in a fixed number of independent Bernoulli
trials, each with the same probability of success. It is used when there are two possible outcomes (success
or failure) in a fixed number of trials, and the trials are independent of each other. For example, it can be
used to model the number of heads in 10 flips of a fair coin.

### Q9. What is the Poisson distribution and where is it applied?
ANS:- The Poisson distribution models the number of events occurring in a fixed interval of time or space, given
that these events happen independently and at a constant rate. It’s commonly used to model rare events,
such as the number of car accidents at a particular intersection in a day or the number of emails received in
an hour.

### Q10. What is a continuous uniform distribution?
ANS:- A continuous uniform distribution describes a situation where all outcomes in a given range are equally
likely. For example, if you randomly pick a number between 0 and 1, every number in that range has an equal
probability of being selected. It’s often visualized as a rectangular shape in probability plots.

### Q11. What are the characteristics of a normal distribution?
ANS:- A normal distribution is a continuous probability distribution that is symmetric about its mean. The
characteristics include;
> The distribution is bell-shaped.
> 
> It is defined by two parameters: the mean (µ) and the standard deviation (σ).
> 
> About 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations,
and 99.7% within three standard deviations.

### Q12. What is the standard normal distribution, and why is it important?
ANS:- The standard normal distribution is a special case of the normal distribution where the mean is 0 and the
standard deviation is 1. It is important because it provides a common reference for comparing different
normal distributions. Values in a standard normal distribution are referred to as Z-scores.

### Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
ANS :-The Central Limit Theorem states that, for a large enough sample size, the sampling distribution of the
sample mean will be approximately normally distributed, regardless of the original distribution of the data.
This is crucial because it allows us to make inferences about a population from a sample, even when the
population distribution is not normal.

### Q14. How does the Central Limit Theorem relate to the normal distribution?
ANS:- The CLT is closely related to the normal distribution because it tells us that the distribution of the sample
mean will approach a normal distribution as the sample size increases. This is true even if the original data
distribution is not normal, making the normal distribution an essential tool in statistical inference.

### Q15.What is the application of Z statistics in hypothesis testing?
ANS:-Z-statistics is used to test hypotheses about population means or proportions when the population standard deviation is known. It's commonly applied in large-sample scenarios (n ≥ 30). Z-tests help determine if a sample differs significantly from a population. Examples include one-sample and two-sample Z-tests.

### Q16.How do you calculate a Z-score, and what does it represent?
ANS:-A Z-score is calculated as Z equals the difference between the observed value (X) and the population mean (μ), divided by the population standard deviation (σ). , where X is the value, μ is the mean, and σ is the standard deviation. It shows how many standard deviations a value is from the mean. Positive Z means above the mean; negative means below. It helps in comparing different data points.

### Q17.What are point estimates and interval estimates in statistics?
ANS:-A point estimate gives a single value to estimate a population parameter, like the sample mean. An interval estimate gives a range with a confidence level (e.g., 95%) in which the parameter likely falls. Point estimates are precise but less reliable alone. Interval estimates express uncertainty better.

### Q18. What is the significance of confidence intervals in statistical analysis?
ANS:- Confidence intervals provide a range of values within which we are confident that the true population
parameter lies. It helps quantify the uncertainty in our estimates and provides more context than a point
estimate alone. A 95% confidence interval, for example, means we are 95% confident that the true value lies
within that interval.

### Q19. What is the relationship between a Z-score and a confidence interval?
ANS:- The Z-score plays a key role in constructing confidence intervals. For a given confidence level, the Z-score
represents the number of standard deviations away from the mean. It’s used in the formula to calculate the
margin of error for the confidence interval, which gives the range around the sample estimate.

### Q20. How are Z-scores used to compare different distributions?
ANS:- Z-scores standardize different data sets so they can be compared on the same scale. By converting
different distributions to Z-scores, we can compare their relative positions and spread, even if the original
distributions have different means and standard deviations.

### Q21. What are the assumptions for applying the Central Limit Theorem?
ANS:-The assumptions for applying the Central Limit Theorem include6
> The sample size should be large enough (typically n > 30).
> 
> The samples should be independent of each other.
> 
> The data should be drawn from the same population.

### Q22. What is the concept of expected value in a probability distribution?
ANS:- The expected value (or mean) of a probability distribution is the long-term average of all possible outcomes,
weighted by their probabilities. It represents the center of the distribution and provides a measure of the
“central tendency” of the distribution.

### Q23. How does a probability distribution relate to the expected outcome of a random variable?
ANS:- A probability distribution shows the likelihood of each possible outcome of a random variable. The expected
value is essentially the weighted average of these outcomes, representing the most likely outcome of the
random variable in the long run.

### Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values from a discrete uniform distribution
samples = np.random.randint(1, 7, size=1000)

# Plot the probability mass function (PMF)
values, counts = np.unique(samples, return_counts=True)
pmf = counts / len(samples)

plt.bar(values, pmf, width=0.5, color='blue', alpha=0.7)
plt.title('Discrete Uniform Distribution PMF')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.show()
```

### Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

def bernoulli_pdf(p, size=1000):
    # p is the probability of success
    x = [0, 1]
    pdf = [stats.bernoulli.pmf(k, p) for k in x]

    return x, pdf

# Parameters
p = 0.7  # probability of success

x, pdf = bernoulli_pdf(p)
plt.bar(x, pdf, color='green', alpha=0.6)
plt.title('Bernoulli Distribution PDF')
plt.xlabel('Outcome (0 or 1)')
plt.ylabel('Probability')
plt.show()
```

### Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random values from a discrete uniform distribution
samples = np.random.randint(1, 7, size=1000)

# Calculate CDF
values, counts = np.unique(samples, return_counts=True)
cdf = np.cumsum(counts) / len(samples)

# Plot CDF
plt.plot(values, cdf, marker='o', color='red')
plt.title('Discrete Uniform Distribution CDF')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.show()
```

### Q9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

# Simulate some data from a normal distribution
data = np.random.normal(0, 1, 1000)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='pink')
plt.title('Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.show()
```

### Q15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

# Sample data
sample_data = np.random.normal(50, 10, 100)

# Parameters for population
population_mean = 50
population_std = 10

# Calculate sample mean and standard error
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)
sample_size = len(sample_data)

# Z-statistic
z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))
print(f"Z-statistic: {z_stat}")

# p-value for two-tailed test
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"P-value: {p_value}")
```

### Q16.Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

# Sample data
sample_data = np.random.normal(50, 10, 100)

# Calculate sample mean and standard error
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data)
sample_size = len(sample_data)

# Confidence interval (95% confidence)
confidence_level = 0.95
confidence_interval = stats.norm.interval(confidence_level, loc=sample_mean, scale=sample_std/np.sqrt(sample_size))

print(f"95% Confidence Interval: {confidence_interval}")
```

### Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

# Parameters for normal distribution
mu, sigma = 50, 10
sample_size = 100

# Generate data
data = np.random.normal(mu, sigma, sample_size)

# Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data)
standard_error = sample_std / np.sqrt(sample_size)

# Confidence interval (95%)
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)
print(f"95% Confidence Interval for the mean: {confidence_interval}")
```

### Q23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.normal(0, 1, 1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_normalized_data = (data - mean) / std

# Plot original and normalized data
plt.hist(data, bins=30, alpha=0.7, label='Original Data')
plt.hist(z_normalized_data, bins=30, alpha=0.7, label='Z-score Normalized Data')
plt.legend()
plt.title('Z-score Normalization')
plt.show()
```
